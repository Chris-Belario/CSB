```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

- Gameboy is the first handheld console released by Nintendo, capable of playing many beloved games such as Tetris and Super Mario

```python
df_w = df.copy()
```

```python
console_count = df_w['Manufacturer'].value_counts()
print(console_count)
```

```python
sales_summary = df.groupby('Manufacturer')['Sales'].sum().reset_index()
sales_summary_desc = sales_summary.sort_values(by='Sales', ascending=False)
print(sales_summary_desc)
```

- The Top 3 Manufacturers with most units sold are Nintendo, Sony, and Microsoft

# GRAPH of Total Units Sold Worldwide

```python
chart = sales_summary_desc.plot(kind='bar',
                        figsize=(10,6))
chart.bar_label(chart.containers[0], label_type='edge')
plt.title('Console Sales by Manufacturers',
          fontfamily = 'Palatino Linotype',
          fontsize=12,
          fontweight='bold')
plt.xlabel('Manufacturers Index')
chart.set_xticklabels(sales_summary_desc.index, rotation=0)
plt.ylabel('Units Sold in Millions')
plt.show()
```
